## Vectors and matrixes

```python
# Вектори та матриці
vector_a = np.arange(6, 11) # numpy array from 1 to 9
vector_b = np.array([1, 2, 3, 4, 5])

print(vector_a * 2)
print(np.log(vector_a))

matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
print(matrix_a)
print(matrix_a.T)




```

```python
# Операції з векторами та матрицями
dot_prod = np.dot(vector_a, vector_b)
print(dot_prod)

matrix_product = np.matmul(matrix_a, matrix_b) # можна й dot, але для матриць краще це, більша точність
print(matrix_product)
```

## 2D plot

```python
# Графік
x_nums = np.linspace(-5, 5, 100)
y_nums = x_nums ** 2

plt.plot(x_nums, y_nums)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Графік функції x^2')
plt.show()
```

## 3D plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_values = np.linspace(-5, 5, 100)
y_values = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x_values, y_values)
z = x**2/3 - y**2/2 - 2

ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Тривимірний графік функції x**2/3 - y**2/2 - 2')
plt.show()
```

```python
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, label='3D spiral')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.legend()

plt.show()
```
